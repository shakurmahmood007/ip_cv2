Capturing a video

```python
import cv2
cap= cv2.VideoCapture("D:\\Avengers-Endgame-Special-Look.mp4")
print(cap)
while True:
    ret, frame= cap.read()
    frame= cv2.resize(frame, (1200,700))
    cv2.imshow("avengers",frame)
    k=cv2.waitKey(2)
    if k==ord("q") and 0xFF:
        break
cap.release()
cv2.destroyAllWindows()
```

video to gray scale

```python
import cv2
cap= cv2.VideoCapture("D:\\Avengers-Endgame-Special-Look.mp4")
print(cap)
while True:
    ret, frame= cap.read()
    frame= cv2.resize(frame, (1200,700))
    gray_scale= cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow("avengers_gs",gray_scale)
    k=cv2.waitKey(8)
    if k==ord("q") and 0xFF:
        break
cap.release()
cv2.destroyAllWindows()
```

opening a webcam video

```python
import cv2
cap= cv2.VideoCapture(0)  # 1 for external webcam
print(cap)
while cap.isOpened():
    ret, frame= cap.read()
    if ret==True:
        frame= cv2.resize(frame, (1200,700))
        cv2.imshow("avengers",frame)
        k=cv2.waitKey(8)
        if k==ord("q") and 0xFF:
            break
cap.release()
cv2.destroyAllWindows()
```

saving a webcam video

```python
import cv2
cap= cv2.VideoCapture(0)  # 1 for external webcam
print(cap)
fourcc = cv2.VideoWriter_fourcc(*"mp4v")  # XVID, DIVX, MJPG, X264, WMV1, WMV2, mp4v
output = cv2.VideoWriter("D:\\Mywebcam.mp4", fourcc, 20, (640, 480))
while cap.isOpened():
    ret, frame= cap.read()
    if ret==True:
        cv2.imshow("Mywebcam",frame)
        output.write(frame)
        k=cv2.waitKey(8)
        if k==ord("q") and 0xFF:
            break
cap.release()
output.release()
cv2.destroyAllWindows()
```

saving a webcam video in grayscale

```python
import cv2
cap= cv2.VideoCapture(0)  # 1 for external webcam
print(cap)
fourcc = cv2.VideoWriter_fourcc(*"XVID")  # XVID, DIVX, MJPG, X264, WMV1, WMV2, mp4v
output = cv2.VideoWriter("D:\\Mywebcam.avi", fourcc, 20, (640, 480), 0)
while cap.isOpened():
    ret, frame= cap.read()
    frame= cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    if ret==True:
        cv2.imshow("Mywebcam",frame)
        output.write(frame)
        k=cv2.waitKey(8)
        if k==ord("q") and 0xFF:
            break
cap.release()
output.release()
cv2.destroyAllWindows()
```

saving a webcam flip video

```python
import cv2
cap= cv2.VideoCapture(0)  # 1 for external webcam
print(cap)
fourcc = cv2.VideoWriter_fourcc(*"mp4v")  # XVID, DIVX, MJPG, X264, WMV1, WMV2, mp4v
output = cv2.VideoWriter("D:\\Mywebcam.mp4", fourcc, 20, (640, 480))
while cap.isOpened():
    ret, frame= cap.read()
    if ret==True:
        frame= cv2.flip(frame, 0) # 0, 1, -1
        cv2.imshow("Mywebcam",frame)
        output.write(frame)
        k=cv2.waitKey(8)
        if k==ord("q") and 0xFF:
            break
cap.release()
output.release()
cv2.destroyAllWindows()
```
